```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set seaborn theme for better-looking plots
sns.set_theme(context='notebook', style='darkgrid', palette='deep')
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Display first few rows to understand the data
iris.head()
```

## Question 1: What is the median petal length for each iris species?

Based on the box plot the  median petal length for each iris species is around 1.6cm for setosa, 4.3cm for versicolor and 5.6cm for virginica.

```python
# Box plot shows the median (middle line in the box)
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='petal_length')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Median Petal Length by Species')
plt.show()
```

## Question 2: What is the mean sepal width for each iris species?

Using the bar graphs the mean sepal width for each iris species are around 3.45 cm for serosa, 2.75 cm for versicolor, and 2.95 cm for virginica.

```python
# Bar plot shows the mean (height of bar)
plt.figure(figsize=(8, 6))
sns.barplot(data=iris, x='species', y='sepal_width', estimator='mean')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.title('Mean Sepal Width by Species')
plt.show()
```

## Question 3: Is there an association between sepal length and width?

When looking at the scatter plot between sepal length and sepal width there is a clear relationship. Each of the species all protray some sort of linear correlation between the two variables indicating that the sepal width and length are not independent.

When investigating setosa species it is clear that it had the strongest linear correlation out of the three species. It also possesed the highest slope out of all the three species indicating that the setosa's width increases much more dramtically as the length increases compared to the two others.

When investigating the versicolor length it is clear that it does have a weaker linear correlation comapred to setosa species. When looking at the scatter plot we can determine that the linear correlation of versicolor is just slightly better than the one for virginica. As some of the points from viriginica are far from the regression line.

```python
# Scatter plot to visualize association between two variables

plt.figure(figsize=(8, 6))
# sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
sns.lmplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    markers=['o', 's', 'D'],
    palette='Set1',
    height=6,
    aspect=1.2,
    legend=False,
    scatter_kws={'s': 60},  # size of scatter points
    line_kws={'linestyle':'--'}
)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Association Between Sepal Length and Width')
plt.legend(title='Species')
plt.show()
```

## Question 4: Compare the distributions of petal width for each species.

When comparing the distribution of petal width for each of the three species it can be seen that the setosa has the lowest variance distribution with most of its points landing near the mean.

When looking at the Versicolor distribution it is clear that its variation is larger than setosa's variation showing a moderate spread from its point across the ran of 0.85 cm to 2.0 cm.

When looking at the virginica distribution it is clear that it has the highest amongst all three variations. It distribution is even distrbuted along its range, protraying a much smoother decline when reaching both ends of its distribution.

```python
# Violin plot shows the distribution of data
plt.figure(figsize=(8, 6))
sns.violinplot(data=iris, x='species', y='petal_width')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Distribution of Petal Width by Species')
plt.show()
```

## Question 5: Are there any petal length outliers?

WHen looking at the box plots we can clearly see that the setosa species was the one that contained the most amount of outliers. Setosa contained outliers from both ends of the spectrum where most of them come from the lower of its range.

When looking at the versicolor species there is only one visible outlier which is at the lower nd of its range.

When looking at the virginica box plot there are clearly no outliters present.

```python
# Box plot shows outliers as individual points beyond the whiskers
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='petal_length')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Petal Length Outliers by Species')
plt.show()
```

## Question 6: Does the association between petal and sepal width change with petal length?

Yes the association between petal and sepal width does change with petal length.

When the petal length are smaller in value there is no association between the petal width and sepal width due to the correlation appearing to be close to a vertical line rather than other function patterns.

When the petal length are 3 or larger there is a clear linear positive association between the petal and sepal width.

```python
# Scatter plot with color coding for petal length
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=iris, x='petal_width', y='sepal_width',
                          hue='petal_length', size='petal_length',
                          sizes=(20, 200), palette='viridis')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Association Between Petal and Sepal Width (colored by Petal Length)')
plt.legend(title='Petal Length', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
